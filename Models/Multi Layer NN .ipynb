# **SINGLE INPUT (Scalar Value)**


```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
layer1 = keras.layers.Dense(units=2,input_shape=[1])
layer2 = keras.layers.Dense(units=1)


model = tf.keras.Sequential([layer1,layer2])
model.compile(optimizer='sgd',loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])

model.fit(xs,ys,epochs=100)

print(model.predict(np.array([10])))
```

```python
print(model.predict(np.array([10])))
```

# **Multiple Input**

```python
layer1 = keras.layers.Dense(units = 2, input_shape = (2,))
layer2 = keras.layers.Dense(units=1)
model = tf.keras.Sequential([layer1,layer2])

opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer = opt, loss = "mean_squared_error")


xs = []
ys = []

for x1 in range(100):
  for x2 in range(100):
    xs.append([x1,x2])
    ys.append([2*x1 - x2 + 2])

xs = np.array(xs)
ys = np.array(ys)


model.fit(xs,ys,epochs=100)
```

```python
print(model.predict([[120.0,-10.0]]))
```
