```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
my_layer = Dense(units=1, input_shape=[1])
model = Sequential([my_layer])
OPT = 'sgd'
LOSS = 'mean_squared_error'
model.compile(optimizer=OPT, loss=LOSS)

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([1.0, 3.0, 5.0, 7.0, 9.0, 11.0])

# Train the model
model.fit(xs, ys, epochs=356)

# Predict
print(model.predict(np.array([10.0])))
```

```python
export_dir = "saved_tf_model/model-1"
tf.saved_model.save(model,export_dir)
```

```python
# Convert tensorFlow model to Tensorflow lite model
converter = tf.lite.TFLiteConverter.from_saved_model(export_dir)
tflite_model = converter.convert()

# Storing Tensorflow lite model onto the disk
import pathlib
tflite_model_file = pathlib.Path("model.tflite")
print("Size of tensorflow lite model in Bytes : ", tflite_model_file.write_bytes(tflite_model))
```
